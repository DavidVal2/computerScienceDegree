```python
import matplotlib.pyplot as plt
import numpy as np


# Valores asociados al número de datos (nd), dato inicial (j) e incremento (d)

nd = 10
j = 10
d = 2



"""
NO MODIFICAR ESTA PARTE DEL CÓDIGO.
"""

a = [j]
x = np.linspace(0, nd-1, nd, dtype = int)
for n in range(1 , len(x)):
  a.append( a[n-1] + d )

fig, ax = plt.subplots(figsize = (4,3) )
ax.scatter(x, a, s = 10)
plt.grid()
plt.title('Comportamiento lineal')
plt.show()
```
