```python
print("1. a ")
from sympy import symbols, oo, limit

n = symbols('n')
expresion = (5*n**2 + 10*n + 3) / (3*n**2 + 15*n + 2)
lim = limit(expresion, n, oo)
display("El límite al infinito de" ,expresion , "es:",lim)

import matplotlib.pyplot as plt
import numpy as np
print("1. b ")
# Definir los valores de n
n = np.arange(1, 210)  # n desde 1 hasta 100

# Calcular los valores de la sucesión 1/n
valores_sucesion = (5*n**2 + 10*n + 3) / (3*n**2 + 15*n + 2)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.scatter(n, valores_sucesion, marker='o', linestyle='-', color='b', label=r'$\frac{1}{n}$')

# Etiquetas y título
plt.xlabel('n')
plt.ylabel(r'$\frac{1}{n}$', rotation=0, labelpad=20)  # Rotar la etiqueta a vertical y ajustar el padding
plt.title('Valores de la Sucesión $\\frac{1}{n}$ a medida que $n$ se incrementa')
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()
print("La grafica se acerca infinitamente hacia 5/3, demostrando la relacion que hay de tamaño entre los 2 algoritmos")
```

```python
print("2. a ")
from sympy import symbols, oo, limit

n = symbols('n')
expresion = (100*n) / (n**2+100)
lim = limit(expresion, n, oo)
display("El límite al infinito de" ,expresion , "es:",lim)

print("2. b ")
import matplotlib.pyplot as plt
import numpy as np
# Definir los valores de n
n = np.arange(1, 1210)  # n desde 1 hasta 100

# Calcular los valores de la sucesión 1/n
valores_sucesion = (100*n) / (n**2+100)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.scatter(n, valores_sucesion, marker='o', linestyle='-', color='b', label=r'$\frac{1}{n}$')

# Etiquetas y título
plt.xlabel('n')
plt.ylabel(r'V', rotation=0, labelpad=20)  # Rotar la etiqueta a vertical y ajustar el padding
plt.title('Velocidad de reparticion de recursos')
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()
print("Al inicio crece rapidamente, pero llega un punto donde es tanta la necesidad de recursos, que la velocidad decrece hasta el infinito.")
```

```python
print("3. a ")
from sympy import symbols, oo, limit, log

n = symbols('n')
expresion = 50 * log(n) - (n/100)
lim = limit(expresion, n, oo)
display("El límite al infinito de" ,expresion , "es:",lim)

print("3. b ")
import matplotlib.pyplot as plt
import numpy as np
# Definir los valores de n
n = np.arange(1, 1111210)  # n desde 1 hasta 100

# Calcular los valores de la sucesión 1/n
valores_sucesion = 50 * np.log(n) - (n/100)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.scatter(n, valores_sucesion, marker='o', linestyle='-', color='b', label=r'$\frac{1}{n}$')

# Etiquetas y título
plt.xlabel('n')
plt.ylabel(r'Rec', rotation=0, labelpad=20)  # Rotar la etiqueta a vertical y ajustar el padding
plt.title('Distribucion de recursos durante gran tiempo')
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()

print("La distribicion de recursos crece durante mucho tiempo, hasta que llega un punto que empieza a decrecer hasta infinito")
```

```python
print("4. D(n) ")
from sympy import symbols, oo, limit, log

n = symbols('n')
expresion = n**2/(log(n))
lim = limit(expresion, n, oo)
display("El límite al infinito de" ,expresion , "es:",lim)

print("Grafica ")
import matplotlib.pyplot as plt
import numpy as np
# Definir los valores de n
n = np.arange(1, 1210)  # n desde 1 hasta 100

# Calcular los valores de la sucesión 1/n
valores_sucesion = n**2/(np.log(n))

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.scatter(n, valores_sucesion, marker='o', linestyle='-', color='b', label=r'$\frac{1}{n}$')

# Etiquetas y título
plt.xlabel('n')
plt.ylabel(r'D', rotation=0, labelpad=20)  # Rotar la etiqueta a vertical y ajustar el padding
plt.title('D(n)')
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()

print("La distribicion de recursos crece exponencialmente en la forma D")


print("4. R(n) ")

n = symbols('n')
expresion = n*log(n)
lim = limit(expresion, n, oo)
display("El límite al infinito de" ,expresion , "es:",lim)

print("Grafica ")

# Definir los valores de n
n = np.arange(1, 1210)  # n desde 1 hasta 100

# Calcular los valores de la sucesión 1/n
valores_sucesion = n*np.log(n)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.scatter(n, valores_sucesion, marker='o', linestyle='-', color='b', label=r'$\frac{1}{n}$')

# Etiquetas y título
plt.xlabel('n')
plt.ylabel(r'R', rotation=0, labelpad=20)  # Rotar la etiqueta a vertical y ajustar el padding
plt.title('R(n)')
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()

print("La distribicion de recursos crece de manera mucho mas reducida en la forma R")
print("Si se quiere ayudar lo mayor posible, lo adecuado es usar la forma D y para mejorar aun, se deberia buscar un algoritmo totalmente exponencial sin restricciones de log.")
```
